```python
#!/usr/bin/python3

#import sklearn to access data for iris flower
from sklearn.datasets import load_iris

#to plot scattered graph
from matplotlib import pyplot as plt

#loading all data
iris=load_iris()

#printing features name
print("Feature Names :",iris.feature_names)

#target names
print("Target Names :",iris.target_names)

# The indices of the features that we are plotting
x_index = 0 #0 is for sepal length
y_index = 3 #3 is for petal width

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)]) 

plt.figure(figsize=(8, 8))#size of the graph
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()

# The indices of the features that we are plotting
x1_index = 1 #0 is for sepal width
y1_index = 2 #3 is for petal length

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)]) 

plt.figure(figsize=(8, 8))#size of the graph
plt.scatter(iris.data[:, x1_index], iris.data[:, y1_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x1_index])
plt.ylabel(iris.feature_names[y1_index])

plt.tight_layout()
plt.show()






```
